```python
from pylab import *
import math
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
xs = range(1,20000)
ys = [math.log10(x) for x in xs]
figure(figsize=(40,4), dpi=400)
plot(xs, ys, linewidth=1.0)
xlabel('n')
ylabel('log10 of n')
title('Log (base-10) of n')
show()
```
